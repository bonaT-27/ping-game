**Cleaning Data**


```python
data['year']
```

year has many non-year values

```python
data=data[data['year'].str.isnumeric()]
```

```python
data['year'].unique()
```

*year is in object; change to intiger*

```python
data['year']=data['year'].astype(int)
```

```python
data['year']
```

```python
data['Price']
```

*Price has Ask for Price*

```python
data=data[data['Price']!='Ask For Price']
```

```python
data['Price']
```

*Price has commas in its prices and is in object*

```python
data['Price'] = data['Price'].fillna('').astype(str).str.replace(' ', '', regex=False)
```

```python
data['Price']=data['Price'].str.replace(',','').astype(int)
```

```python
data['Price']
```

```python
data['kms_driven']
```

kms_driven has object values with kms at last.

```python
data['kms_driven']=data['kms_driven'].str.split().str.get(0).str.replace(',','')
```

It has nan values and two rows have 'Petrol' in them

```python
data=data[data['kms_driven'].str.isnumeric()]
```

```python
data['kms_driven']=data['kms_driven'].astype(int)
```

fuel_type has nan values

```python
data=data[~data['fuel_type'].isna()]
```

```python
data.shape
```

```python
data.info()
```

```python
data['name']
```

Company does not need any cleaning now. Changing car names. Keeping only the first three words

```python
data['name']=data['name'].str.split().str.slice(start=0,stop=3).str.join(' ')
```

```python
data['name']
```

```python
data
```

```python
data=data.reset_index(drop=True)
```

**Cleaned data**

```python
data
```

```python
data=data[data['Price']<6e6].reset_index(drop=True)
```

**Training Data**

```python
x=data.drop(columns='Price')
y=data['Price']
```

```python
x
```

Target column

```python
y
```

Applying Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

Creating an OneHotEncoder object to contain all the possible categories

```python
ohe = OneHotEncoder()
ohe.fit(x[['name', 'company', 'fuel_type']])
```

Creating a column transformer to transform categorical columns

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```


Linear Regression Model

```python
lr=LinearRegression()
```


Making a pipeline

```python
pipe=make_pipeline(column_trans,lr)
```


Fitting the model

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
y_pred
```

Checking R2 Score

```python
r2_score(y_test,y_pred)
```

Finding the model with a random state of TrainTestSplit where the model was found to give almost 0.2 as r2_score

```python
score=[]
for i in range(1000):
  x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
  lr=LinearRegression()
  pipe=make_pipeline(column_trans,lr)
  pipe.fit(x_train,y_train)
  y_pred=pipe.predict(x_test)
  # Append the calculated r2_score to the 'score' list
  score.append(r2_score(y_test,y_pred))
```

```python
np.argmax(score)
```

```python
score[np.argmax(score)]
```
